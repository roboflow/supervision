```python
import cv2


def blur(frame, i):
    return cv2.GaussianBlur(frame, (11, 11), 0)


sv.Video("gemist.mp4").save("blurred.mp4", callback=blur, show_progress=True)
```

```python
sv.Video("gemist.mp4").save(
    "timelapse.mp4", fps=60, callback=lambda f, i: f, show_progress=True
)
```
